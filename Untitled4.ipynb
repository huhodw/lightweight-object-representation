```python
import torch
import torch.nn as nn
import torch.nn.functional as F

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Device: {torch.cuda.get_device_name(0)}")
```

```python
class SlotAttention(nn.Module):
    def __init__(self, num_slots, dim, iters=3, eps=1e-8, hidden_dim=128):
        super().__init__()
        self.num_slots = num_slots
        self.iters = iters
        self.eps = eps
        self.scale = dim ** -0.5

        self.slots_mu = nn.Parameter(torch.randn(1, 1, dim))
        self.slots_log_sigma = nn.Parameter(torch.zeros(1, 1, dim))

        self.to_q = nn.Linear(dim, dim, bias=False)
        self.to_k = nn.Linear(dim, dim, bias=False)
        self.to_v = nn.Linear(dim, dim, bias=False)

        self.gru = nn.GRUCell(dim, dim)

        self.mlp = nn.Sequential(
            nn.Linear(dim, hidden_dim),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_dim, dim),
        )

        self.norm_input = nn.LayerNorm(dim)
        self.norm_slots = nn.LayerNorm(dim)
        self.norm_mlp = nn.LayerNorm(dim)

    def forward(self, inputs, return_attn=False):
        B, N, D = inputs.shape

        mu = self.slots_mu.expand(B, self.num_slots, -1)
        sigma = self.slots_log_sigma.exp().expand(B, self.num_slots, -1)
        slots = mu + sigma * torch.randn_like(mu)

        inputs = self.norm_input(inputs)
        k = self.to_k(inputs)
        v = self.to_v(inputs)

        attn = None
        for _ in range(self.iters):
            slots_prev = slots
            slots = self.norm_slots(slots)

            q = self.to_q(slots)

            attn_logits = torch.einsum('bkd,bnd->bkn', q, k) * self.scale
            attn = attn_logits.softmax(dim=1) + self.eps

            attn_weights = attn / attn.sum(dim=-1, keepdim=True)
            updates = torch.einsum('bkn,bnd->bkd', attn_weights, v)

            slots = self.gru(
                updates.reshape(-1, D),
                slots_prev.reshape(-1, D)
            ).reshape(B, self.num_slots, D)
            slots = slots + self.mlp(self.norm_mlp(slots))

        if return_attn:
            return slots, attn
        return slots

print("SlotAttention クラスを定義しました")
```

```python
class SoftPositionEmbed(nn.Module):
    """座標情報を特徴マップに足し込む"""
    def __init__(self, hidden_dim, resolution):
        super().__init__()
        self.embedding = nn.Linear(4, hidden_dim, bias=True)
        self.register_buffer('grid', self._build_grid(resolution))

    def _build_grid(self, resolution):
        H, W = resolution
        ranges = [torch.linspace(0.0, 1.0, steps=r) for r in resolution]
        grid_y, grid_x = torch.meshgrid(ranges[0], ranges[1], indexing='ij')
        grid = torch.stack([grid_x, 1.0 - grid_x, grid_y, 1.0 - grid_y], dim=-1)
        return grid.unsqueeze(0)

    def forward(self, inputs):
        pos_emb = self.embedding(self.grid)
        pos_emb = pos_emb.permute(0, 3, 1, 2)
        return inputs + pos_emb

print("SoftPositionEmbed クラスを定義しました")
```

```python
class Encoder(nn.Module):
    """画像 (B, 3, H, W) → ピクセル系列 (B, N, D) に変換するエンコーダ"""
    def __init__(self, resolution, hidden_dim=64):
        super().__init__()
        self.conv1 = nn.Conv2d(3, hidden_dim, kernel_size=5, padding=2)
        self.conv2 = nn.Conv2d(hidden_dim, hidden_dim, kernel_size=5, padding=2)
        self.conv3 = nn.Conv2d(hidden_dim, hidden_dim, kernel_size=5, padding=2)
        self.conv4 = nn.Conv2d(hidden_dim, hidden_dim, kernel_size=5, padding=2)

        self.pos_embed = SoftPositionEmbed(hidden_dim, resolution)

        self.mlp = nn.Sequential(
            nn.LayerNorm(hidden_dim),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_dim, hidden_dim),
        )

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = F.relu(self.conv4(x))

        x = self.pos_embed(x)

        x = x.flatten(start_dim=2)
        x = x.transpose(1, 2)

        x = self.mlp(x)

        return x

print("Encoder クラスを定義しました")
```

```python
class Decoder(nn.Module):
    """スロット1個から (4, H, W) の画像 + マスクを生成"""
    def __init__(self, hidden_dim=64, decoder_init_resolution=(8, 8)):
        super().__init__()
        self.decoder_init_resolution = decoder_init_resolution

        self.pos_embed = SoftPositionEmbed(hidden_dim, decoder_init_resolution)

        self.deconv1 = nn.ConvTranspose2d(
            hidden_dim, hidden_dim,
            kernel_size=5, stride=2, padding=2, output_padding=1
        )
        self.deconv2 = nn.ConvTranspose2d(
            hidden_dim, hidden_dim,
            kernel_size=5, stride=2, padding=2, output_padding=1
        )
        self.conv3 = nn.Conv2d(hidden_dim, hidden_dim, kernel_size=5, padding=2)
        self.conv_out = nn.Conv2d(hidden_dim, 4, kernel_size=3, padding=1)

    def forward(self, slots):
        B, K, D = slots.shape

        x = slots.reshape(B * K, D)

        H0, W0 = self.decoder_init_resolution
        x = x.unsqueeze(-1).unsqueeze(-1)
        x = x.expand(-1, -1, H0, W0)

        x = self.pos_embed(x)

        x = F.relu(self.deconv1(x))
        x = F.relu(self.deconv2(x))
        x = F.relu(self.conv3(x))

        x = self.conv_out(x)

        x = x.reshape(B, K, 4, x.shape[-2], x.shape[-1])

        recons = x[:, :, :3, :, :]
        masks = x[:, :, 3:4, :, :]

        return recons, masks

print("Decoder クラスを定義しました")
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

torch.manual_seed(42)
decoder = Decoder(hidden_dim=64, decoder_init_resolution=(8, 8)).to(device)

dummy_slots = torch.randn(2, 7, 64).to(device)
print(f"入力スロットの形状: {dummy_slots.shape}")

recons, masks = decoder(dummy_slots)
print(f"再構成画像の形状: {recons.shape}")
print(f"マスクの形状:     {masks.shape}")
print(f"期待される形状:   (2, 7, 3, 32, 32) と (2, 7, 1, 32, 32)")

print(f"\n再構成画像の値範囲: [{recons.min().item():.3f}, {recons.max().item():.3f}]")
print(f"マスクの値範囲:     [{masks.min().item():.3f}, {masks.max().item():.3f}]")

total_params = sum(p.numel() for p in decoder.parameters() if p.requires_grad)
print(f"\nDecoder パラメータ数: {total_params:,}")
```

```python
class SlotAttentionAutoEncoder(nn.Module):
    """画像 → スロット → 再構成画像 の全体モデル"""
    def __init__(self, resolution=(32, 32), num_slots=7, num_iterations=3, hidden_dim=64):
        super().__init__()
        self.resolution = resolution
        self.num_slots = num_slots

        # エンコーダ
        self.encoder = Encoder(resolution=resolution, hidden_dim=hidden_dim)

        # SlotAttention
        self.slot_attention = SlotAttention(
            num_slots=num_slots,
            dim=hidden_dim,
            iters=num_iterations
        )

        # デコーダ
        self.decoder = Decoder(
            hidden_dim=hidden_dim,
            decoder_init_resolution=(8, 8)
        )

    def forward(self, image, return_slots=False):
        # image: (B, 3, H, W)
        B = image.shape[0]

        # ① エンコード: (B, 3, H, W) → (B, N, D)
        features = self.encoder(image)

        # ② Slot Attention: (B, N, D) → (B, K, D)
        slots = self.slot_attention(features)

        # ③ デコード: (B, K, D) → (B, K, 3, H, W) と (B, K, 1, H, W)
        recons, masks = self.decoder(slots)

        # ④ α合成: マスクをスロット軸でsoftmax、RGB を重み付き平均
        masks_softmax = masks.softmax(dim=1)  # (B, K, 1, H, W)
        recon_combined = (recons * masks_softmax).sum(dim=1)  # (B, 3, H, W)

        if return_slots:
            return recon_combined, recons, masks_softmax, slots
        return recon_combined

print("SlotAttentionAutoEncoder クラスを定義しました")
```

```python
torch.manual_seed(42)
model = SlotAttentionAutoEncoder(
    resolution=(32, 32),
    num_slots=7,
    num_iterations=3,
    hidden_dim=64
).to(device)

# ダミー画像
dummy_image = torch.randn(2, 3, 32, 32).to(device)
print(f"入力画像の形状: {dummy_image.shape}")

# forward
recon_combined, recons, masks_softmax, slots = model(dummy_image, return_slots=True)

print(f"\n最終再構成画像: {recon_combined.shape}")
print(f"各スロット画像: {recons.shape}")
print(f"各スロットマスク: {masks_softmax.shape}")
print(f"スロット: {slots.shape}")

# マスクが正しく確率分布になっているか確認
mask_sum = masks_softmax.sum(dim=1)  # スロット軸で合計
print(f"\nマスク合計の範囲（1.0 になるべき）: [{mask_sum.min().item():.4f}, {mask_sum.max().item():.4f}]")

# 学習可能パラメータ数
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\n全体のパラメータ数: {total_params:,}")

# 再構成損失の計算テスト（これが訓練で使う損失関数）
recon_loss = F.mse_loss(recon_combined, dummy_image)
print(f"再構成損失（MSE）: {recon_loss.item():.4f}")

# 逆伝播のテスト
recon_loss.backward()
print("逆伝播成功")
```

```python
import numpy as np
from torch.utils.data import Dataset, DataLoader

class ShapesDataset(Dataset):
    """ランダムに円を配置した32×32の画像を生成するデータセット"""
    def __init__(self, num_samples=10000, image_size=32, max_objects=3, seed=None):
        self.num_samples = num_samples
        self.image_size = image_size
        self.max_objects = max_objects
        # 色パレット（RGB、0〜1 の範囲）
        self.colors = [
            [1.0, 0.2, 0.2],   # 赤
            [0.2, 1.0, 0.2],   # 緑
            [0.2, 0.2, 1.0],   # 青
            [1.0, 1.0, 0.2],   # 黄
            [1.0, 0.2, 1.0],   # マゼンタ
            [0.2, 1.0, 1.0],   # シアン
        ]
        # 再現性のためのランダム生成器
        self.rng = np.random.default_rng(seed)

    def __len__(self):
        return self.num_samples

    def __getitem__(self, idx):
        # 白背景の画像 (3, H, W)
        image = np.ones((3, self.image_size, self.image_size), dtype=np.float32)

        # 物体数をランダムに決める（2〜max_objects個）
        num_objs = self.rng.integers(2, self.max_objects + 1)

        for _ in range(num_objs):
            # ランダムな中心位置と半径を決める
            radius = self.rng.integers(3, 7)
            cx = self.rng.integers(radius, self.image_size - radius)
            cy = self.rng.integers(radius, self.image_size - radius)
            color = self.colors[self.rng.integers(0, len(self.colors))]

            # 円を描画（マスクを作って色を塗る）
            y_grid, x_grid = np.meshgrid(
                np.arange(self.image_size),
                np.arange(self.image_size),
                indexing='ij'
            )
            mask = (y_grid - cy) ** 2 + (x_grid - cx) ** 2 <= radius ** 2

            for c in range(3):
                image[c, mask] = color[c]

        # torch.Tensor に変換して返す
        return torch.from_numpy(image)

print("ShapesDataset クラスを定義しました")
```

```python
import matplotlib.pyplot as plt

# データセットを作る
torch.manual_seed(42)
dataset = ShapesDataset(num_samples=10000, image_size=32, seed=42)
print(f"データセットのサイズ: {len(dataset)}")

# 8サンプルを取得して可視化
fig, axes = plt.subplots(1, 8, figsize=(16, 2))
for i in range(8):
    image = dataset[i]  # (3, 32, 32)
    # matplotlibで表示するため (H, W, C) に並び替え
    image_np = image.permute(1, 2, 0).numpy()
    axes[i].imshow(image_np)
    axes[i].set_title(f"Sample {i}")
    axes[i].axis('off')
plt.tight_layout()
plt.show()

# 1サンプルの形状を確認
sample = dataset[0]
print(f"\n1サンプルの形状: {sample.shape}")
print(f"値の範囲: [{sample.min().item():.3f}, {sample.max().item():.3f}]")
print(f"データ型: {sample.dtype}")
```

```python
# 訓練用データセットとデータローダー
train_dataset = ShapesDataset(num_samples=10000, image_size=32, seed=42)
train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True,        # 毎エポック順序をシャッフル
    num_workers=2,       # 並列読み込み
    pin_memory=True,     # GPU転送を高速化
)

# 検証用データセット（別のseedで、訓練と被らないサンプル）
val_dataset = ShapesDataset(num_samples=500, image_size=32, seed=123)
val_loader = DataLoader(
    val_dataset,
    batch_size=64,
    shuffle=False,       # 検証はシャッフル不要
    num_workers=2,
    pin_memory=True,
)

print(f"訓練サンプル数: {len(train_dataset)}")
print(f"検証サンプル数: {len(val_dataset)}")
print(f"訓練バッチ数（1エポックあたり）: {len(train_loader)}")
print(f"検証バッチ数: {len(val_loader)}")

# 1バッチ取り出して形状確認
batch = next(iter(train_loader))
print(f"\n1バッチの形状: {batch.shape}")
print(f"期待される形状: (64, 3, 32, 32)")
print(f"値の範囲: [{batch.min().item():.3f}, {batch.max().item():.3f}]")
print(f"デバイス: {batch.device}")
```

```python
# ハイパーパラメータ
NUM_SLOTS = 7        # スロット数（物体の最大個数より多めに）
HIDDEN_DIM = 64      # 隠れ次元
NUM_ITERATIONS = 3   # Slot Attention の反復回数
LEARNING_RATE = 4e-4 # 学習率
WARMUP_STEPS = 1000  # 学習率ウォームアップのステップ数
NUM_STEPS = 3000     # 総訓練ステップ数
GRAD_CLIP = 1.0      # 勾配クリッピングの最大ノルム

# モデル作成
torch.manual_seed(42)
model = SlotAttentionAutoEncoder(
    resolution=(32, 32),
    num_slots=NUM_SLOTS,
    num_iterations=NUM_ITERATIONS,
    hidden_dim=HIDDEN_DIM
).to(device)

# オプティマイザ
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

# 学習率ウォームアップ用のラムダ関数
def lr_lambda(step):
    if step < WARMUP_STEPS:
        return step / WARMUP_STEPS  # 0 → 1 に線形に上げる
    else:
        return 1.0                   # その後は一定

scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=lr_lambda)

print(f"モデル作成完了")
print(f"パラメータ数: {sum(p.numel() for p in model.parameters()):,}")
print(f"訓練ステップ数: {NUM_STEPS}")
print(f"バッチサイズ: 64")
print(f"合計処理サンプル数: {NUM_STEPS * 64:,}")
```

```python
def train_step(model, batch, optimizer, scheduler, grad_clip):
    """1バッチを処理する関数"""
    model.train()  # 訓練モード（dropoutなどを有効化）

    # バッチをGPUに転送
    batch = batch.to(device)

    # 1. 勾配をクリア
    optimizer.zero_grad()

    # 2. forward: 再構成画像を計算
    recon = model(batch)

    # 3. 損失計算（MSE: 平均二乗誤差）
    loss = F.mse_loss(recon, batch)

    # 4. backward: 勾配を計算
    loss.backward()

    # 5. 勾配クリッピング
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=grad_clip)

    # 6. パラメータ更新
    optimizer.step()

    # 7. 学習率スケジューラを1ステップ進める
    scheduler.step()

    return loss.item()

print("train_step 関数を定義しました")
```

```python
# 動作確認: 20ステップだけ訓練
print("=== 動作確認: 20ステップだけ訓練 ===\n")

step = 0
losses_test = []
data_iter = iter(train_loader)

for step in range(20):
    try:
        batch = next(data_iter)
    except StopIteration:
        data_iter = iter(train_loader)
        batch = next(data_iter)

    loss = train_step(model, batch, optimizer, scheduler, GRAD_CLIP)
    losses_test.append(loss)

    current_lr = optimizer.param_groups[0]['lr']
    print(f"Step {step+1:3d} | Loss: {loss:.4f} | LR: {current_lr:.6f}")

print("\n動作確認完了")
```

```python
def visualize_reconstruction(model, val_loader, num_samples=4, step=None):
    """訓練途中の再構成画像を可視化"""
    model.eval()

    with torch.no_grad():  # 勾配計算を無効化（メモリ節約）
        batch = next(iter(val_loader))[:num_samples].to(device)
        recon_combined, recons, masks, slots = model(batch, return_slots=True)

    # 図の準備：入力、再構成、各スロットのマスク
    K = model.num_slots
    fig, axes = plt.subplots(num_samples, 2 + K, figsize=(2 * (2 + K), 2 * num_samples))

    for i in range(num_samples):
        # 入力画像
        input_img = batch[i].cpu().permute(1, 2, 0).numpy()
        axes[i, 0].imshow(input_img)
        axes[i, 0].set_title("Input" if i == 0 else "")
        axes[i, 0].axis('off')

        # 再構成画像
        recon_img = recon_combined[i].cpu().permute(1, 2, 0).numpy()
        recon_img = np.clip(recon_img, 0, 1)  # 値を [0, 1] に収める
        axes[i, 1].imshow(recon_img)
        axes[i, 1].set_title("Recon" if i == 0 else "")
        axes[i, 1].axis('off')

        # 各スロットのマスク
        for k in range(K):
            mask_k = masks[i, k, 0].cpu().numpy()
            axes[i, 2 + k].imshow(mask_k, cmap='gray', vmin=0, vmax=1)
            axes[i, 2 + k].set_title(f"Slot {k}" if i == 0 else "")
            axes[i, 2 + k].axis('off')

    if step is not None:
        plt.suptitle(f"Step {step}", y=1.02)
    plt.tight_layout()
    plt.show()

print("visualize_reconstruction 関数を定義しました")
```

```python
import time
import os
import json
import torch

# ============================================
# 訓練設定の確認
# ============================================
print("=" * 60)
print("【訓練前のチェック】")
print("=" * 60)

# 必要な変数が定義されているか確認
required_vars = ['model', 'optimizer', 'scheduler', 'train_loader', 'val_loader',
                 'NUM_STEPS', 'GRAD_CLIP', 'train_step']
for var_name in required_vars:
    exists = var_name in dir() or var_name in globals()
    status = "✅" if exists else "❌"
    print(f"  {status} {var_name}")

print(f"\n  NUM_STEPS = {NUM_STEPS}")
print(f"  GRAD_CLIP = {GRAD_CLIP}")
print(f"  optimizer の lr = {optimizer.param_groups[0]['lr']}")

# Driveをマウント（保存に備える）
from google.colab import drive
drive.mount('/content/drive')

save_dir = '/content/drive/MyDrive/slot_attention_phase1'
os.makedirs(save_dir, exist_ok=True)
print(f"\n  保存先確認: {save_dir} ✅")

print("\n" + "=" * 60)
print("3秒後に訓練開始...")
print("=" * 60)
time.sleep(3)

# ============================================
# 訓練ループ本体
# ============================================
print(f"\n=== 本番訓練開始（{NUM_STEPS} ステップ）===\n")

train_losses = []
data_iter = iter(train_loader)
start_time = time.time()

# 中間保存のタイミング（500ステップごと）
SAVE_EVERY = 500

for step in range(NUM_STEPS):
    try:
        batch = next(data_iter)
    except StopIteration:
        data_iter = iter(train_loader)
        batch = next(data_iter)

    loss = train_step(model, batch, optimizer, scheduler, GRAD_CLIP)
    train_losses.append(loss)

    # 100ステップごとに進捗表示
    if (step + 1) % 100 == 0:
        elapsed = time.time() - start_time
        avg_loss = sum(train_losses[-100:]) / len(train_losses[-100:])
        current_lr = optimizer.param_groups[0]['lr']
        print(f"Step {step+1:4d}/{NUM_STEPS} | "
              f"Loss: {avg_loss:.4f} | "
              f"LR: {current_lr:.6f} | "
              f"Time: {elapsed:.1f}s")

    # ★ 500ステップごとに「中間保存」（重要）
    if (step + 1) % SAVE_EVERY == 0:
        ckpt_path = os.path.join(save_dir, f'checkpoint_step{step+1}.pth')
        torch.save({
            'step': step + 1,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'scheduler_state_dict': scheduler.state_dict(),
            'train_losses': train_losses,
            'loss': loss,
        }, ckpt_path)
        print(f"  💾 中間保存: checkpoint_step{step+1}.pth (loss={loss:.4f})")

# ============================================
# 訓練完了
# ============================================
total_time = time.time() - start_time
print(f"\n=== 訓練完了 ===")
print(f"総訓練時間: {total_time:.1f}秒 ({total_time/60:.1f}分)")
print(f"最終損失: {train_losses[-1]:.4f}")

# ============================================
# 最終保存（複数ファイル）
# ============================================
print("\n=== 最終結果を保存中... ===")

# 1. モデル重み（メインの保存）
final_weights_path = os.path.join(save_dir, 'model_weights.pth')
torch.save(model.state_dict(), final_weights_path)
size_mb = os.path.getsize(final_weights_path) / (1024 * 1024)
print(f"✅ 重み保存: {final_weights_path} ({size_mb:.2f} MB)")

# 2. 訓練損失
losses_path = os.path.join(save_dir, 'train_losses.json')
with open(losses_path, 'w') as f:
    json.dump(train_losses, f)
print(f"✅ 損失履歴: {losses_path}")

# 3. メタ情報（再現性のため）
meta = {
    'num_steps': NUM_STEPS,
    'final_loss': train_losses[-1],
    'min_loss': min(train_losses),
    'training_time_min': total_time / 60,
    'total_params': sum(p.numel() for p in model.parameters()),
    'lr_initial': optimizer.param_groups[0]['lr'],  # 注：訓練後の値かも
}
meta_path = os.path.join(save_dir, 'meta.json')
with open(meta_path, 'w') as f:
    json.dump(meta, f, indent=2)
print(f"✅ メタ情報: {meta_path}")

# ============================================
# 訓練成功の即時検証（保存後すぐ）
# ============================================
print("\n=== 保存直後の動作確認 ===")
model.eval()
device = next(model.parameters()).device
with torch.no_grad():
    test_images = next(iter(val_loader))[:4].to(device)
    recon, recons, masks, _ = model(test_images, return_slots=True)
    test_mse = ((recon - test_images) ** 2).mean().item()

    # スロット間の差を確認
    slot_means = [masks[:, k].mean().item() for k in range(masks.shape[1])]
    slot_std = torch.tensor(slot_means).std().item()

print(f"  検証MSE  : {test_mse:.5f}")
print(f"  スロットstd: {slot_std:.5f}")

if test_mse < 0.05 and slot_std > 0.01:
    print(f"  ✅ 訓練成功！M5に進めます")
else:
    print(f"  ⚠️ 何か問題がある可能性、結果を確認してください")

print("\n" + "=" * 60)
print("🎉 全工程完了")
print("=" * 60)
print("\n保存されたファイル一覧:")
for f in sorted(os.listdir(save_dir)):
    full_path = os.path.join(save_dir, f)
    size_kb = os.path.getsize(full_path) / 1024
    print(f"  {f:40s}  {size_kb:>10.1f} KB")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

# ============================================
# 1. データ取得
# ============================================
NUM_SAMPLES = 6  # 表示するサンプル数

model.eval()
device = next(model.parameters()).device

with torch.no_grad():
    # 検証データから取得
    images = next(iter(val_loader))
    images = images[:NUM_SAMPLES].to(device)

    recon_combined, recons, masks, slots = model(images, return_slots=True)

B, K, _, H, W = recons.shape
print(f"Batch={B}, Slots={K}, Image={H}x{W}")

# ============================================
# 2. CPU & numpy 変換
# ============================================
images_np   = images.cpu().numpy()         # [B, 3, H, W]
recon_np    = recon_combined.cpu().numpy() # [B, 3, H, W]
recons_np   = recons.cpu().numpy()         # [B, K, 3, H, W]
masks_np    = masks.cpu().numpy()          # [B, K, 1, H, W]

# 各スロットの「α合成済み画像」（黒背景での貢献度）
slot_images = recons_np * masks_np         # [B, K, 3, H, W]

# ============================================
# 3. ヘルパー関数
# ============================================
def to_display_rgb(img):
    """[3, H, W] → [H, W, 3] でクリップ"""
    img = np.clip(img, 0, 1)
    return img.transpose(1, 2, 0)

def to_display_mask(mask):
    """[1, H, W] → [H, W] でクリップ"""
    return np.clip(mask[0], 0, 1)

# ============================================
# 4. 可視化（2段構成）
# 上段: Input | Recon | Slot1合成 | Slot2合成 | ... | Slot7合成
# 下段: 空白  | 空白   | Slot1マスク | Slot2マスク | ... | Slot7マスク
# ============================================
n_cols = 2 + K  # Input + Recon + K slots
fig, axes = plt.subplots(NUM_SAMPLES * 2, n_cols,
                          figsize=(n_cols * 1.4, NUM_SAMPLES * 2 * 1.4))

col_titles = ['Input', 'Recon'] + [f'Slot {k+1}' for k in range(K)]

for b in range(NUM_SAMPLES):
    row_rgb = b * 2       # RGB行
    row_mask = b * 2 + 1  # マスク行

    # === RGB行 ===
    # Input
    axes[row_rgb, 0].imshow(to_display_rgb(images_np[b]))
    axes[row_rgb, 0].axis('off')
    if b == 0:
        axes[row_rgb, 0].set_title('Input', fontsize=10)

    # Recon
    axes[row_rgb, 1].imshow(to_display_rgb(recon_np[b]))
    axes[row_rgb, 1].axis('off')
    if b == 0:
        axes[row_rgb, 1].set_title('Recon', fontsize=10)

    # 各スロット（α×RGB）
    for k in range(K):
        axes[row_rgb, 2 + k].imshow(to_display_rgb(slot_images[b, k]))
        axes[row_rgb, 2 + k].axis('off')
        if b == 0:
            axes[row_rgb, 2 + k].set_title(f'Slot {k+1}', fontsize=10)

    # === マスク行 ===
    # Input/Recon列は空白
    axes[row_mask, 0].axis('off')
    axes[row_mask, 1].axis('off')
    axes[row_mask, 0].text(0.5, 0.5, 'Mask↓', ha='center', va='center',
                            transform=axes[row_mask, 0].transAxes, fontsize=8, color='gray')

    # 各スロットのマスクのみ（グレースケール）
    for k in range(K):
        axes[row_mask, 2 + k].imshow(to_display_mask(masks_np[b, k]),
                                      cmap='gray', vmin=0, vmax=1)
        axes[row_mask, 2 + k].axis('off')

plt.tight_layout()

# 保存
save_path = '/content/drive/MyDrive/slot_attention_phase1/m5_visualization.png'
plt.savefig(save_path, dpi=120, bbox_inches='tight')
plt.show()

print(f"\n✅ 可視化保存完了: {save_path}")
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# ============================================
# 多くのバッチをスキャンして「再構成誤差が大きい」サンプルを探す
# ============================================
model.eval()
device = next(model.parameters()).device

errors = []  # (error, image, recon, recons, masks)

print("=== 失敗ケース探索中... ===")
with torch.no_grad():
    for batch_idx, images in enumerate(val_loader):
        images = images.to(device)
        recon, recons, masks, slots = model(images, return_slots=True)

        # サンプルごとの再構成誤差を計算
        per_sample_mse = ((recon - images) ** 2).mean(dim=[1, 2, 3])  # [B]

        for i in range(images.shape[0]):
            errors.append({
                'mse': per_sample_mse[i].item(),
                'image': images[i].cpu().numpy(),
                'recon': recon[i].cpu().numpy(),
                'recons': recons[i].cpu().numpy(),
                'masks': masks[i].cpu().numpy(),
            })

        if batch_idx >= 5:  # 6バッチ分（約400サンプル）で十分
            break

# 誤差順にソート
errors.sort(key=lambda x: x['mse'], reverse=True)
print(f"スキャン完了: {len(errors)} サンプル")
print(f"最大誤差: {errors[0]['mse']:.4f}")
print(f"最小誤差: {errors[-1]['mse']:.4f}")
print(f"平均誤差: {np.mean([e['mse'] for e in errors]):.4f}")

# ============================================
# Worst 4 と Best 4 を可視化
# ============================================
def to_display(img):
    img = np.clip(img, 0, 1)
    if img.ndim == 3:
        img = img.transpose(1, 2, 0)
    return img

def plot_cases(case_list, title, save_name):
    K = case_list[0]['recons'].shape[0]
    n_cases = len(case_list)
    n_cols = 2 + K

    fig, axes = plt.subplots(n_cases, n_cols, figsize=(n_cols * 1.4, n_cases * 1.4))
    if n_cases == 1:
        axes = axes[np.newaxis, :]

    for i, case in enumerate(case_list):
        axes[i, 0].imshow(to_display(case['image']))
        axes[i, 0].axis('off')
        axes[i, 0].set_ylabel(f"MSE={case['mse']:.3f}", fontsize=8)
        if i == 0:
            axes[i, 0].set_title('Input', fontsize=9)

        axes[i, 1].imshow(to_display(case['recon']))
        axes[i, 1].axis('off')
        if i == 0:
            axes[i, 1].set_title('Recon', fontsize=9)

        for k in range(K):
            slot_img = case['recons'][k] * case['masks'][k]
            axes[i, 2+k].imshow(to_display(slot_img))
            axes[i, 2+k].axis('off')
            if i == 0:
                axes[i, 2+k].set_title(f'Slot {k+1}', fontsize=9)

    plt.suptitle(title, fontsize=12, y=1.01)
    plt.tight_layout()
    save_path = f'/content/drive/MyDrive/slot_attention_phase1/{save_name}.png'
    plt.savefig(save_path, dpi=120, bbox_inches='tight')
    plt.show()
    print(f"✅ 保存: {save_path}")

# ワースト4ケース（失敗例）
print("\n=== ワースト4ケース（失敗例）===")
plot_cases(errors[:4], 'Worst 4 Cases (Failure Examples)', 'm5_failures')

# ベスト4ケース（成功例）
print("\n=== ベスト4ケース（成功例）===")
plot_cases(errors[-4:], 'Best 4 Cases (Success Examples)', 'm5_successes')
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# ============================================
# 同じ画像を5回モデルに通して、スロット割り当ての変動を見る
# ============================================
model.eval()
device = next(model.parameters()).device

# 1サンプルを選ぶ
with torch.no_grad():
    images = next(iter(val_loader))
    test_image = images[0:1].to(device)  # [1, 3, H, W]

NUM_TRIALS = 5

# 5回推論
all_recons = []
all_masks = []

with torch.no_grad():
    for trial in range(NUM_TRIALS):
        # ※ Slot Attentionの初期化はモデル内部でランダムに行われる
        recon, recons, masks, slots = model(test_image, return_slots=True)
        all_recons.append(recons[0].cpu().numpy())  # [K, 3, H, W]
        all_masks.append(masks[0].cpu().numpy())    # [K, 1, H, W]

# ============================================
# 可視化: 行=試行, 列=スロット
# ============================================
K = all_recons[0].shape[0]

fig, axes = plt.subplots(NUM_TRIALS + 1, K + 1,
                          figsize=((K+1) * 1.4, (NUM_TRIALS+1) * 1.4))

# 一番上の行: 入力画像
axes[0, 0].imshow(np.clip(test_image[0].cpu().numpy().transpose(1,2,0), 0, 1))
axes[0, 0].set_title('Input', fontsize=9)
axes[0, 0].axis('off')
for k in range(K):
    axes[0, k+1].axis('off')
    axes[0, k+1].set_title(f'Slot {k+1}', fontsize=9)

# 各試行の結果
for trial in range(NUM_TRIALS):
    axes[trial+1, 0].text(0.5, 0.5, f'Trial {trial+1}',
                           ha='center', va='center', fontsize=10,
                           transform=axes[trial+1, 0].transAxes)
    axes[trial+1, 0].axis('off')

    for k in range(K):
        slot_img = all_recons[trial][k] * all_masks[trial][k]
        axes[trial+1, k+1].imshow(np.clip(slot_img.transpose(1,2,0), 0, 1))
        axes[trial+1, k+1].axis('off')

plt.suptitle('Slot Assignment Stability Across Trials', fontsize=12, y=1.01)
plt.tight_layout()

save_path = '/content/drive/MyDrive/slot_attention_phase1/m5_stability.png'
plt.savefig(save_path, dpi=120, bbox_inches='tight')
plt.show()
print(f"✅ 保存: {save_path}")

# ============================================
# 数値的な安定性指標
# ============================================
# 試行間でのマスクの相関を計算
print("\n=== スロット割り当ての安定性 ===")
print("各スロットが試行間でどれくらい同じ領域を担当するか")
print("（1.0 に近いほど安定）\n")

masks_array = np.stack(all_masks, axis=0)  # [trials, K, 1, H, W]
masks_flat = masks_array.reshape(NUM_TRIALS, K, -1)  # [trials, K, H*W]

for k in range(K):
    # 試行間の相関
    correlations = []
    for i in range(NUM_TRIALS):
        for j in range(i+1, NUM_TRIALS):
            mi = masks_flat[i, k]
            mj = masks_flat[j, k]
            if mi.std() > 0 and mj.std() > 0:
                corr = np.corrcoef(mi, mj)[0, 1]
                correlations.append(corr)

    if correlations:
        avg_corr = np.mean(correlations)
        bar = '█' * max(0, int(avg_corr * 30)) if avg_corr > 0 else ''
        print(f"  Slot {k+1}: 平均相関 = {avg_corr:+.3f}  {bar}")
    else:
        print(f"  Slot {k+1}: 計算不可（マスクが定数）")
```
